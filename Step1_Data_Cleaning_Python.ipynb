# 1. Procedure

- In order to clean the dataset efficiently and effectively, I will only focus on cleaning the datasets that are essential for my analysis to understand how consumers are using their smart device.
- Datasets to focus: *__dailyActivity_merged.csv__*, *__sleepDay_merged.csv__*, and *__weightLogInfo_merged.csv__*

## 1.1 dailyActivity_merged.csv
- dailyActivity has 940 observations, no missing values, no duplicate rows.
- The daily activity contains records between __12 Apr-12 May 2016__.
- There are 33 distinct users in this dataset and 21 of them use the smart device consecutively for 31 days.
- Converted Id data type and Activity Date data type to 'string' and 'datetime', respectively.
- Added a Day column and TotalMinutes column.

```python
dailyActivity.info()
```

```python
dailyActivity.isnull().sum()
```

```python
# Check duplicates
dailyActivity.duplicated().sum()
```

```python
# Check duplicates on Id and ActivityDate for any duplicate entries per Id
dailyActivity.duplicated(subset=['Id','ActivityDate']).sum()
```

```python
dailyActivity
```

```python
# Find earliest date of record
dailyActivity.sort_values(by='ActivityDate', ascending=True, inplace=False)
```

```python
# Find latest date of record
dailyActivity.sort_values(by='ActivityDate', ascending=False, inplace=False)
```

```python
# Check distinct Id columns
dailyActivity['Id'].value_counts()
```

```python
filter_user = dailyActivity[dailyActivity['Id'] == 1503960366]
filter_user
```

```python
# Convert Id data type from 'int64' to 'string' and ActivityDate data type from 'string' to 'datetime'
dailyActivity['Id'] = dailyActivity['Id'].astype('string')
dailyActivity['ActivityDate'] = pd.to_datetime(dailyActivity['ActivityDate'])
dailyActivity.info()
```

```python
dailyActivity['ActivityDate'].dt.day
```

```python
filter_user = dailyActivity[dailyActivity['Id'] == '1503960366']
filter_user
```

```python
# Add Day column and TotalMinutes column
dailyActivity['Day'] = dailyActivity['ActivityDate'].dt.day_name()
dailyActivity['TotalMinutes'] = dailyActivity['VeryActiveMinutes'] + dailyActivity['FairlyActiveMinutes'] + dailyActivity['LightlyActiveMinutes'] + dailyActivity['SedentaryMinutes']
```

```python
dailyActivity.info()
```

```python
# Export the cleaned dailyActivity dataset
dailyActivity.to_csv("dataset/dailyActivity_clean.csv", index=False)
```
